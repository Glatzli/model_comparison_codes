# CAP depth

```python
height_seegrube = 2278
p_seg = height_to_pressure_std(height_seegrube * units('m')).magnitude
```

```python
#find nearest value in df to a given value
import numpy as np
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

#### WRF_ACINN

```python
def wrf_acinn(df):
    a = df['temperature'].where(df['pressure'] == np.max(df['pressure'])).values
    wrf_T_surf = a[~np.isnan(a)]
    wrf_T_max = np.max(df["temperature"].values)
    b = df['temperature'].where(df['pressure'] == find_nearest(df['pressure'], value=p_seg)).values
    wrf_T_seg = b[~np.isnan(b)]

    c = df['pressure'].where(df['temperature'] == wrf_T_max).values
    wrf_p_of_T_max = c[~np.isnan(c)]
    wrf_p_surf = np.max(df['pressure']).values

    wrf_z_surf = np.min(df.z).values
    d = df.z.where(df['temperature'] == wrf_T_max).values
    wrf_z_of_T_max = d[~np.isnan(c)]
    
    return wrf_T_surf, wrf_T_max, wrf_T_seg, wrf_p_of_T_max, wrf_p_surf, wrf_z_surf, wrf_z_of_T_max

wrf_k_T_surf, wrf_k_T_max, wrf_k_T_seg, wrf_k_p_of_T_max, wrf_k_p_surf, wrf_k_z_surf, wrf_k_z_of_T_max = wrf_acinn(df_wrf_k)
wrf_i_T_surf, wrf_i_T_max, wrf_i_T_seg, wrf_i_p_of_T_max, wrf_i_p_surf, wrf_i_z_surf, wrf_i_z_of_T_max = wrf_acinn(df_wrf_i)
```

#### AROME

```python
def arome(df):
    a = df['temperature'].where(df['pressure'] == np.max(df['pressure'])).values
    arome_T_surf = a[~np.isnan(a)]
    arome_T_max = np.max(df["temperature"].values)
    b = df['temperature'].where(df['pressure'] == find_nearest(df['pressure'], value=p_seg)* units('hPa')).values
    arome_T_seg = b[~np.isnan(b)]

    c = df['pressure'].where(df['temperature'] == find_nearest(df['temperature'], value=arome_T_max)* units('degree_Celsius')).values
    arome_p_of_T_max = c[~np.isnan(c)]
    arome_p_surf = np.max(df['pressure']).values

    arome_z_surf = np.min(df.height).values
    d = df.height.where(df['temperature'] == find_nearest(df['temperature'], value=arome_T_max)* units('degree_Celsius')).values
    arome_z_of_T_max = d[~np.isnan(c)]
    
    return arome_T_surf, arome_T_max, arome_T_seg, arome_p_of_T_max, arome_p_surf, arome_z_surf, arome_z_of_T_max

arome_k_T_surf, arome_k_T_max, arome_k_T_seg, arome_k_p_of_T_max, arome_k_p_surf, arome_k_z_surf, arome_k_z_of_T_max = arome(df_arome_k)
arome_i_T_surf, arome_i_T_max, arome_i_T_seg, arome_i_p_of_T_max, arome_i_p_surf, arome_i_z_surf, arome_i_z_of_T_max = arome(df_arome_i)

```

#### UM

```python
def ukmo(df):
    a = df['temperature'].where(df['pressure'] == np.max(df['pressure'])).values
    ukmo_T_surf = a[~np.isnan(a)]
    ukmo_T_max = np.max(df["temperature"].values)
    b = df['temperature'].where(df['pressure'] == find_nearest(df['pressure'], value=p_seg)).values
    ukmo_T_seg = b[~np.isnan(b)]
    
    c = df['pressure'].where(df['temperature'] == ukmo_T_max).values
    ukmo_p_of_T_max = c[~np.isnan(c)]
    ukmo_p_surf = np.max(df['pressure'])

    ukmo_z_surf = np.min(df.geopotential_height)
    d = df.geopotential_height.where(df['temperature'] == ukmo_T_max).values
    ukmo_z_of_T_max = d[~np.isnan(c)]
    
    return ukmo_T_surf, ukmo_T_max, ukmo_T_seg, ukmo_p_of_T_max, ukmo_p_surf, ukmo_z_surf, ukmo_z_of_T_max

ukmo_k_T_surf, ukmo_k_T_max, ukmo_k_T_seg, ukmo_k_p_of_T_max, ukmo_k_p_surf, ukmo_k_z_surf, ukmo_k_z_of_T_max = ukmo(data_ukmo_k)
ukmo_i_T_surf, ukmo_i_T_max, ukmo_i_T_seg, ukmo_i_p_of_T_max, ukmo_i_p_surf, ukmo_i_z_surf, ukmo_i_z_of_T_max = ukmo(data_ukmo_i)
```

# Brunt-Väisälä-Frequency

$N = \sqrt{\frac{g}{\theta} \frac{\delta \theta}{\delta z}}$   with z coordinates

###### Non-dimensional mountain height (Froud-number) H

$H = \frac{N_m h}{u_n}$

Pressure at Seegrube = 800 hPa = 2278.78 m

```python
def bv_freq(theta0, theta1, z0, z1):
    theta_mean = (theta1 - theta0)/2
    N = np.sqrt(g/theta_mean * (theta1-theta0)/(z1-z0))
    return N #* units('1/s')
```

```python
def pot_t(df, df_T_surf, df_p_of_T_max, df_T_max, df_T_seg):
    th_surf = potential_temperature(np.max(df['pressure'])*units('hPa'), df_T_surf*units('degree_Celsius')).magnitude
    th_max = potential_temperature(df_p_of_T_max*units('hPa'), df_T_max*units('degree_Celsius')).magnitude
    th_seg = potential_temperature(p_seg*units('hPa'), df_T_seg*units('degree_Celsius')).magnitude
    return np.array([th_surf, th_max, th_seg])
```

```python
#TH KUFSTEIN
data_k = [df_wrf_k,  data_ukmo_k, df_icon_k, df_icon2TE_k]
T_surf_k = [wrf_k_T_surf, ukmo_k_T_surf, icon_k_T_surf, icon2TE_k_T_surf]
p_of_T_max_k = [wrf_k_p_of_T_max, ukmo_k_p_of_T_max, icon_k_p_of_T_max, icon2TE_k_p_of_T_max]
T_max_k = [wrf_k_T_max, ukmo_k_T_max, icon_k_T_max, icon2TE_k_T_max]
T_seg_k = [wrf_k_T_seg, ukmo_k_T_seg, icon_k_T_seg, icon2TE_k_T_seg]

out = np.zeros([4,3])
for i in range(4):
    for j in range(3):
        out[i][j] = pot_t(data_k[i], T_surf_k[i], p_of_T_max_k[i], T_max_k[i], T_seg_k[i])[j]
    
wrf_th_surf_k, wrf_th_max_k, wrf_th_seg_k = out[0]
ukmo_th_surf_k, ukmo_th_max_k, ukmo_th_seg_k = out[1]
icon_th_surf_k, icon_th_max_k, icon_th_seg_k = out[2]
icon2TE_th_surf_k, icon2TE_th_max_k, icon2TE_th_seg_k = out[3]
```

```python
#TH IMST
data_i = [df_wrf_i,  data_ukmo_i, df_icon_i, df_icon2TE_i]
T_surf_i = [wrf_i_T_surf, ukmo_i_T_surf, icon_i_T_surf, icon2TE_i_T_surf]
p_of_T_max_i = [wrf_i_p_of_T_max, ukmo_i_p_of_T_max, icon_i_p_of_T_max, icon2TE_i_p_of_T_max]
T_max_i = [wrf_i_T_max, ukmo_i_T_max, icon_i_T_max, icon2TE_i_T_max]
T_seg_i = [wrf_i_T_seg, ukmo_i_T_seg, icon_i_T_seg, icon2TE_i_T_seg]

out_i = np.zeros([4,3])
for i in range(4):
    for j in range(3):
        out_i[i][j] = pot_t(data_i[i], T_surf_i[i], p_of_T_max_i[i], T_max_i[i], T_seg_i[i])[j]
    
wrf_th_surf_i, wrf_th_max_i, wrf_th_seg_i = out_i[0]
ukmo_th_surf_i, ukmo_th_max_i, ukmo_th_seg_i = out_i[1]
icon_th_surf_i, icon_th_max_i, icon_th_seg_i = out_i[2]
icon2TE_th_surf_i, icon2TE_th_max_i, icon2TE_th_seg_i = out_i[3]
```

```python
def pot_t_arome(df, df_T_surf, df_p_of_T_max, df_T_max, df_T_seg):
    th_surf = potential_temperature(np.max(df['pressure']), df_T_surf*units('degree_Celsius')).magnitude
    th_max = potential_temperature(df_p_of_T_max*units('hPa'), df_T_max*units('degree_Celsius')).magnitude
    th_seg = potential_temperature(p_seg*units('hPa'), df_T_seg*units('degree_Celsius')).magnitude
    return th_surf, th_max, th_seg

arome_th_surf_k, arome_th_max_k, arome_th_seg_k = pot_t_arome(df_arome_k, arome_k_T_surf, arome_k_p_of_T_max, arome_k_T_max, arome_k_T_seg)
arome_th_surf_i, arome_th_max_i, arome_th_seg_i = pot_t_arome(df_arome_i, arome_i_T_surf, arome_i_p_of_T_max, arome_i_T_max, arome_i_T_seg)

arome_th_surf_k = arome_th_surf_k.item()
arome_th_max_k = arome_th_max_k.item()
arome_th_seg_k = arome_th_seg_k.item()

arome_th_surf_i = arome_th_surf_i.item()
arome_th_max_i = arome_th_max_i.item()
arome_th_seg_i = arome_th_seg_i.item()
```

```python
th_surf_k = np.array([wrf_th_surf_k, arome_th_surf_k, ukmo_th_surf_k, icon_th_surf_k, icon2TE_th_surf_k])
th_max_k = np.array([wrf_th_max_k, arome_th_max_k, ukmo_th_max_k, icon_th_max_k, icon2TE_th_max_k])
th_seg_k = np.array([wrf_th_seg_k, arome_th_seg_k, ukmo_th_seg_k, icon_th_seg_k, icon2TE_th_seg_k])
th_surf_i = np.array([wrf_th_surf_i, arome_th_surf_i, ukmo_th_surf_i, icon_th_surf_i, icon2TE_th_surf_i])
th_max_i = np.array([wrf_th_max_i, arome_th_max_i, ukmo_th_max_i, icon_th_max_i, icon2TE_th_max_i])
th_seg_i = np.array([wrf_th_seg_i, arome_th_seg_i, ukmo_th_seg_i, icon_th_seg_i, icon2TE_th_seg_i])

#bv_freq(th_surf, th_seg, z_surf, z_seg)[0]
bv_freq(th_surf_k, th_max_k, k_z_surf, k_z_max)

bv_seg_k = bv_freq(th_surf_k, th_seg_k, k_z_surf, z_seg)
bv_max_k = bv_freq(th_surf_k, th_max_k, k_z_surf, k_z_max)
bv_seg_i = bv_freq(th_surf_i, th_seg_i, i_z_surf, z_seg)
bv_max_i = bv_freq(th_surf_i, th_max_i, i_z_surf, i_z_max)

brunt_väisälä_freq = pd.DataFrame({'model': ['wrf','arome', 'um','icon','icon2TE'],
                               'bv_freq_seg_surf_k [$s^{-1}$]': bv_freq(th_surf_k, th_seg_k, k_z_surf, z_seg),
                               'bv_freq_maxT_surf_k [$s^{-1}$]': bv_freq(th_surf_k, th_max_k, k_z_surf, k_z_max),
                               'bv_freq_seg_surf_i [$s^{-1}$]': bv_freq(th_surf_i, th_seg_i, i_z_surf, z_seg),
                               'bv_freq_maxT_surf_i [$s^{-1}$]': bv_freq(th_surf_i, th_max_i, i_z_surf, i_z_max),
                                }).set_index('model')
brunt_väisälä_freq
```

# Bulk Temperature Gradient

$\frac{T_1 - T_0}{z_1 - z_0}$

```python
T_surf_k = np.array([wrf_k_T_surf, arome_k_T_surf, ukmo_k_T_surf, icon_k_T_surf, icon2TE_k_T_surf]).T.flatten()
T_surf_i = np.array([wrf_i_T_surf, arome_i_T_surf, ukmo_i_T_surf, icon_i_T_surf, icon2TE_i_T_surf]).T.flatten()

T_max_k = np.array([wrf_k_T_max, arome_k_T_max, ukmo_k_T_max, icon_k_T_max, icon2TE_k_T_max])
T_max_i = np.array([wrf_i_T_max, arome_i_T_max, ukmo_i_T_max, icon_i_T_max, icon2TE_i_T_max])

T_seg_k = np.array([wrf_k_T_seg, arome_k_T_seg, ukmo_k_T_seg, icon_k_T_seg, icon2TE_k_T_seg]).T.flatten()
T_seg_i = np.array([wrf_i_T_seg, arome_i_T_seg, ukmo_i_T_seg, icon_i_T_seg, icon2TE_i_T_seg]).T.flatten()


# Bulk Temperature Gradient between Seegrube and Surface (Temperature in °C, height in m)
(T_seg_k - T_surf_k) / (z_seg - k_z_surf)
(T_seg_i - T_surf_i) / (z_seg - i_z_surf)
# Bulk Temperature Gradient between Max. T and Surface (Temperature in °C, height in m)
(T_max_k - T_surf_k) / (k_z_max - k_z_surf)
(T_max_i - T_surf_i) / (i_z_max - i_z_surf)

bulk_seg_k = (T_seg_k - T_surf_k) / (z_seg - k_z_surf)
bulk_seg_i = (T_seg_i - T_surf_i) / (z_seg - i_z_surf)
bulk_max_k = (T_max_k - T_surf_k) / (k_z_max - k_z_surf)
bulk_max_i = (T_max_i - T_surf_i) / (i_z_max - i_z_surf)
bulk_th_seg_k = (th_seg_k - th_surf_k).T.flatten() / (z_seg - k_z_surf)
bulk_th_seg_i = (th_seg_i - th_surf_i).T.flatten() / (z_seg - i_z_surf)
bulk_th_max_k = (th_max_k - th_surf_k).T.flatten() / (k_z_max - k_z_surf)
bulk_th_max_i = (th_max_i - th_surf_i).T.flatten() / (i_z_max - i_z_surf)


bulk_temp_grad = pd.DataFrame({'model': ['wrf','arome','um','icon','icon2TE'],
                               'bulk_T_grad_seg_surf_k [$°C*m^{-1}$]': (T_seg_k - T_surf_k) / (z_seg - k_z_surf),
                               'bulk_T_grad_seg_surf_i [$°C*m^{-1}$]': (T_seg_i - T_surf_i) / (z_seg - i_z_surf),
                               'bulk_T_grad_max_surf_k [$°C*m^{-1}$]': (T_max_k - T_surf_k) / (k_z_max - k_z_surf),
                               'bulk_T_grad_max_surf_i [$°C*m^{-1}$]': (T_max_i - T_surf_i) / (i_z_max - i_z_surf),
                                }).set_index('model')

bulk_temp_grad_th = pd.DataFrame({'model': ['wrf','arome','um','icon','icon2TE'],
                               'bulk_theta_grad_seg_surf_k [$K*m^{-1}$]': (th_seg_k - th_surf_k).T.flatten() / (z_seg - k_z_surf),
                               'bulk_theta_grad_seg_surf_i [$K*m^{-1}$]': (th_seg_i - th_surf_i).T.flatten() / (z_seg - i_z_surf),
                               'bulk_theta_grad_max_surf_k [$K*m^{-1}$]': (th_max_k - th_surf_k).T.flatten() / (k_z_max - k_z_surf),
                               'bulk_theta_grad_max_surf_i [$K*m^{-1}$]': (th_max_i - th_surf_i).T.flatten() / (i_z_max - i_z_surf),
                                }).set_index('model')
display(bulk_temp_grad)
display(bulk_temp_grad_th)


```

# Valley Heat Deficit

use absolute temperature T by definition

```python
#bulk valley heat deficit

def valley_heat_def(T0, T1, z0, z1):
    gamma = -0.0098 #* units('K*m**(-1)')
    rho_0 = 1.2 #* units('kg*m**(-3)')
    c_p = 1005 #* units('J*kg**(-1)*K**(-1)')
    H = 0.5 * rho_0 * c_p * (z1-z0)**2 * ((T1-T0)/(z1-z0) + abs(gamma))
    return H * units('J*m**(-2)')
```

```python
# bulk valley heat deficit between Seegrube and Surface (Temperature in °C, not pot.T!, height in z)
#valley_heat_def(T_surf, T_seg, z_surf, z_seg)
# bulk valley heat deficit between Seegrube and Surface (Temperature in °C, not pot.T!, height in z)
#valley_heat_def(T_surf, T_max, z_surf, z_max)

H_seg_k= valley_heat_def(T_surf_k, T_seg_k, k_z_surf, z_seg).magnitude
H_seg_i= valley_heat_def(T_surf_i, T_seg_i, i_z_surf, z_seg).magnitude
H_max_k= valley_heat_def(T_surf_k, T_max_k, k_z_surf, k_z_max).magnitude
H_max_i= valley_heat_def(T_surf_i, T_max_i, i_z_surf, i_z_max).magnitude

valley_heat_deficit = pd.DataFrame({'H_seg_surf_k $[Jm^{-2}]$': valley_heat_def(T_surf_k, T_seg_k, k_z_surf, z_seg),
                                    'H_seg_surf_i $[Jm^{-2}]$': valley_heat_def(T_surf_i, T_seg_i, i_z_surf, z_seg),
                                    'H_max_surf_k $[Jm^{-2}]$': valley_heat_def(T_surf_k, T_max_k, k_z_surf, k_z_max),
                                    'H_max_surf_i $[Jm^{-2}]$': valley_heat_def(T_surf_i, T_max_i, i_z_surf, i_z_max),
                                    'model': ['wrf','arome','um','icon','icon2TE']
                                    }).set_index('model')
valley_heat_deficit
```

```python
#find nearest value in df to a given value
import numpy as np
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
def find_nearest_and_index(data, nearest):
    for index, value in enumerate(data):
        if value == find_nearest(data, nearest):
            return(index)
```
